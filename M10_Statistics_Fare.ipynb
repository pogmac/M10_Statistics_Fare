```python
# manipulacja danymi
import numpy as np
import pandas as pd

# wizualizacja danych
import matplotlib.pyplot as plt
import seaborn as sns

# statystyczna analiza danych
from scipy import stats

# przygotowanie danych
from sklearn.preprocessing import StandardScaler, MinMaxScaler
dataset = pd.read_csv('titanic.csv')
dataset
```

```python
# Checking the number of non null observations:
fare_values = dataset.loc[dataset['Fare'].notnull(), 'Fare'].values
print(fare_values.shape)
#print(fare_values)
```

```python
plt.hist(fare_values, bins = 100)
plt.show()
```

```python
mean_fare = np.round(np.mean(fare_values))
print(mean_fare)
```

```python
plt.hist(fare_values, bins = 100)
plt.axvline(x =mean_fare, color = 'red',label ='Average')
plt.legend(loc='upper right')
plt.show()
```

```python
median_fare = np.median(fare_values)
print(median_fare)
```

```python
plt.hist(fare_values, bins = 100)
plt.axvline(x =mean_fare, color = 'red',label ='Average')
plt.axvline(x =median_fare, color = 'green',label ='Median')
plt.legend(loc='upper right')
plt.show()
```

```python
mode_fare = stats.mode(fare_values, keepdims = True)
print(mode_fare)
```

```python
q0 = np.quantile(fare_values, 0.0)
q1 = np.round(np.quantile(fare_values, 0.25))
q3 = np.round(np.quantile(fare_values, 0.75))
q4 = np.round(np.quantile(fare_values, 1.0))
print(f'Q0: {q0}')
print(f'Q1: {q1}')
print(f'Q2: {median_fare}')
print(f'Q3: {q3}')
print(f'Q4: {q4}')
```

```python
plt.hist(fare_values, bins = 100)
plt.axvline(x =mean_fare, color = 'red',label ='Average')
plt.axvline(x =median_fare, color = 'green',label ='Median')
plt.axvline(x = mode_fare[0], color='orange', label = 'Mode')
plt.axvline(x = q0, color='black', label = 'Min')
plt.axvline(x = q1, color='black', label = 'Q1')
plt.axvline(x = q3, color='black', label = 'Q3')
plt.axvline(x = q4, color='black', label = 'Max')
plt.legend(loc='upper right')
plt.show()
```

```python
range_fare = max(fare_values) - min(fare_values)
print(range_fare)
```

```python
#inter quantile range
iqr = q3 - q1
print(iqr)
```

```python
plt.boxplot(fare_values)
plt.show()
```

```python
variance_fare = np.var(fare_values, ddof=1)
print(variance_fare)
```

```python
print(variance_fare**(1/2))
```

```python
standard_deviation_fare =np.std(fare_values, ddof =1)
print(standard_deviation_fare)
```

```python
# Skalowanie zmiennej / Variable Scaling
# Standaryzacja / standardization
```

```python
scaler_standardized = StandardScaler()
standardized_fare_values = scaler_standardized.fit_transform(fare_values.reshape(-1, 1))
plt.hist(standardized_fare_values, bins=100)
plt.axvline(x = standardized_fare_values.mean(), color='red', label = 'Mean')
plt.show()
print(f'Mean: {standardized_fare_values.mean()}')
print(f'Standard Deviation: {np.std(standardized_fare_values, ddof=1)}')
```

```python
scaler_MinMax = MinMaxScaler()
normalized_fare_values = scaler_MinMax.fit_transform(fare_values.reshape(-1, 1))
plt.hist(normalized_fare_values, bins=100)
plt.axvline(x = normalized_fare_values.mean(), color='red', label = 'Mean')
plt.show()
print(f'Mean: {normalized_fare_values.mean()}')
print(f'Standard Deviation: {np.std(normalized_fare_values, ddof=1)}')
```
